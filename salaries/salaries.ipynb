# POLINMOMIAL REGRRESION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:,-1].values
```

# TRAINING LINEAR REGRRESOR MODEL ON WHOLE DATA SET 

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x,y)
```

# TRAINING POLYNOMIAL REGRESSION MODEL ON WHOLE DATA SET

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg= PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(x)
regressor_2=LinearRegression()
regressor_2.fit(x_poly,y)
```

# VISULAISING THE LINEAR REGRESSION MODEL

```python
plt.scatter(x,y,color='red')
plt.plot(x,regressor.predict(x), color='blue')
plt.title('truth or bluff(linear regrresion)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

# VISUALISING THE POLYINOMIAL REGRESSION MODEL

```python
plt.scatter(x,y,color='red')
plt.plot(x,regressor_2.predict(x_poly), color='blue')
plt.title('truth or bluff(POLYNOMIAL regrresion)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

# PREDICITING A NEW RESULT BY LINEAR REGRESSION

```python
regressor.predict([[6.5]])
```

# PREDICITING A NEW RESULT BY POLYNOMIAL REGRESSION

```python
regressor_2.predict(poly_reg.fit_transform([[6.5]]))
```
